__중심극한 정리 시뮬__

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
mother = np.random.rand(20000)
sns.distplot(mother)
```

```python
def center_(sample_size):
    res = []
    for _ in range(2000):
        res.append(np.mean(np.random.choice(mother, sample_size)))
    return res
```

```python
sns.distplot(center_(12))
```
